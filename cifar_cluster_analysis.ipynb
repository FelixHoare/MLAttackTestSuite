```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cifar_cluster_data = pd.read_csv('data/cifar10_clustermatch_results.csv')
cifar_cluster_data.head()
```

```python
model_0 = cifar_cluster_data[cifar_cluster_data['Model'] == 0]
model_1 = cifar_cluster_data[cifar_cluster_data['Model'] == 1]
model_2 = cifar_cluster_data[cifar_cluster_data['Model'] == 2]
model_3 = cifar_cluster_data[cifar_cluster_data['Model'] == 3]
model_4 = cifar_cluster_data[cifar_cluster_data['Model'] == 4]
model_5 = cifar_cluster_data[cifar_cluster_data['Model'] == 5]
```

```python
model_0_pr_05 = model_0[model_0['Poison rate'] == 0.5]
model_0_pr_1 = model_0[model_0['Poison rate'] == 1.0]
model_0_pr_2 = model_0[model_0['Poison rate'] == 2.0]
```

```python
model_0_pr_1['Target Damage'] = model_0['Clean Model, Clean Subpopulation accuracy (subpop baseline)'] - model_0['Poisoned Model, Clean Subpopulation accuracy (target)']
model_0_pr_1['Collateral Damage'] = model_0['Clean Model Accuracy'] - model_0['Poisoned Model, Clean Test Data accuracy (collateral)']

m0_highest1_target = model_0_pr_1.nlargest(1, 'Target Damage')
m0_highest5_target = model_0_pr_1.nlargest(5, 'Target Damage')
m0_highest10_target = model_0_pr_1.nlargest(10, 'Target Damage')

m0_highest_1_collat = m0_highest1_target['Collateral Damage']
m0_highest_5_collat = m0_highest5_target['Collateral Damage']
m0_highest_10_collat = m0_highest10_target['Collateral Damage']

print(f"Highest individual target damage for model 0: {m0_highest1_target['Target Damage'].mean()}")
print(f"Highest 5 target damage for model 0: {m0_highest5_target['Target Damage'].mean()}")
print(f"Highest 10 target damage for model 0: {m0_highest10_target['Target Damage'].mean()}")

print(f"Highest individual collateral damage for model 0: {m0_highest_1_collat.mean()}")
print(f"Highest 5 collateral damage for model 0: {m0_highest_5_collat.mean()}")
print(f"Highest 10 collateral damage for model 0: {m0_highest_10_collat.mean()}")
```
